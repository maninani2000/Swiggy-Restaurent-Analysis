```python
# Initialize an empty list
html_list = []

# Loop through the elements and get their innerHTML
for ele in elements:
    html = ele.get_attribute("innerHTML")
    # Append the innerHTML to the list
    html_list.append(html)

# Print the list of innerHTML
print(html_list)
```

```python
import re
import pandas as pd

# List to store the restaurant data
restaurant_data = []

# Loop through the html strings and extract the restaurant data
for html in html_list:
#     restaurant_ids = re.findall('data-rest-id="(.*?)"', html)
    restaurant_links = re.findall('<a href="(.*?)">', html)
    name = re.findall('<div class="nA6kb">(.*?)</div>', html)
    rating = re.findall('<span class="icon-star _537e4"></span><span>(.*?)</span>', html)
    cusins = re.findall('<div class="_1gURR" title="(.*?)">', html)
    price = re.findall('<div class="nVWSi">(.*?)</div>', html)
    delivery_time = re.findall('<div>(\d+) MINS</div>', html)
    
    # If price or cusins not found, set it to empty string
    if not price:
        price = [""]
    if not cusins:
        cusins = [""]
        
    # Add the restaurant data to the list
    restaurant_data.append([restaurant_links[0],name[0], rating[0], cusins[0], price[0], delivery_time[0]])
    restaurant_data.append([restaurant_links[1],name[1], rating[1], cusins[1], price[1], delivery_time[1]])
    restaurant_data.append([restaurant_links[2],name[2], rating[2], cusins[2], price[2], delivery_time[2]])
#     restaurant_data.append([restaurant_links[3],name[3], rating[3], cusins[3], price[3], delivery_time[3]])


# Create a dataframe from the restaurant data
df88 = pd.DataFrame(restaurant_data, columns=['Link','Restaurant Name', 'Rating', 'Cusins', 'Price for Two', 'Delivery Time'])
df88
```
